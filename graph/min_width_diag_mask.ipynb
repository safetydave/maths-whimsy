#Minimum width diagonal mask

Creates a diagonal mask of a certain width, even into the corners

```python
import numpy as np
```

```python
def diagonal_mask(array_size, min_width):
    mask_base = np.ones((array_size, array_size), dtype=bool)
    half_width = min_width // 2
    mask_u = np.triu(mask_base, -half_width)
    mask_l = np.tril(mask_base, half_width)
    mask_diagonal = np.logical_and(mask_u, mask_l)
    mask_diagonal[:half_width, :min_width] = True
    mask_diagonal[-half_width:, -min_width:] = True
    return mask_diagonal
```

```python
print(diagonal_mask(8, 3))
```

```python
print(diagonal_mask(12, 7))
```
